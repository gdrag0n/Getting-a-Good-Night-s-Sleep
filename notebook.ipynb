## ðŸ’¾ The data: sleep_health_data.csv

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months. The data is saved as `sleep_health_data.csv`.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

```python
import pandas as pd
```

```python
df=pd.read_csv('sleep_health_data.csv', header=0, sep=",")
print(df.shape)
print(df)
```

## Task 1
# Checking for Occupation with lowest Average Sleep Duration

Creating dictionary with Occupations associated by their list of Sleep duration arrays

```python
A='Occupation'
B='Sleep Duration'

d_sd={}
for i in range(len(A)):
    
    occ=df.at[i,A]
    sd=df.at[i,B]
    
    #sd=int(str(sd))
    
    if occ in d_sd:
        d_sd[occ].append(sd)
    else:
        #d[occ]=sd
        d_sd[occ]=[sd]          #creates a list with the first value
                                #ensuring next values get appended easily
        
print(d_sd)
```

With given sleep durations, it is required to calculate average

```python
l=[]
for i in d_sd:
    avg=sum(d_sd[i])/len(d_sd[i])
    l.append(avg)
```

After averages, find occupation with minimum average sleep duration

```python
l_sd=l.index(min(l))   #index of lowest average value
#print(l_sd)

#retrive occ at index from key list: list(d_sd.keys())
lowest_sleep_occ=list(d_sd.keys())[l_sd] 
print(lowest_sleep_occ)
```

## Task 2
# Checking for Occupation with lowest Avg Sleep Quality

```python
A='Occupation'
B='Quality of Sleep'

d_qos={}
for i in range(len(A)):
    
    occ=df.at[i,A]
    qos=df.at[i,B]
    
    qos=int(str(qos))
    
    if occ in d_qos:
        d_qos[occ].append(qos)
    else:
        #d_qos[occ]=qos
        d_qos[occ]=[qos]        #creates a list with the first value
                            #ensuring next values get appended easily
        
print(d_qos)
```

With given sleep quality, it is required to calculate average

```python
l=[]
for i in d_qos:
    avg=sum(d_qos[i])/len(d_qos[i])
    l.append(avg)
```

find occupation with lowest average qos

```python
l_qos=l.index(min(l))   #index of lowest average value
#print(l_qos)

#retrive occ at index from key list: list(d_sd.keys())
lowest_sleep_quality_occ=list(d_qos.keys())[l_qos] 
print(lowest_sleep_quality_occ)
```

## Task 3
# BMI

```python
A='BMI Category'

counts = df[A].value_counts().to_dict()
print(counts)
```

```python
A='BMI Category'
B='Sleep Disorder'

d_dis={'Normal':0,
      'Overweight':0,
      'Obese':0}

n=df.shape[0]
for i in range(n):
    
    bmi=df.at[i,A]
    dis=df.at[i,B]
    
    if dis=='Insomnia':
        d_dis[bmi]+=1
    else:
        continue
        
print(d_dis)
```

```python
for i in d_dis:
    d_dis[i]=round(d_dis[i]/counts[i], 2)

print(d_dis)
```
